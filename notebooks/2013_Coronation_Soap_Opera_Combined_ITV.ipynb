```python
# Read the June 15th power data from the data folder.
import pandas as pd
from pickup_utilities import *
df = pd.read_csv("../data/complete.csv")
df['tstamp'] = df.apply(tstampCalc, axis = 1)
```

```python
df.head()
```

```python
# The 1900 refers to the time

df_Coronation_July_2_1930 = df.loc[df['day']==2].loc[df['month']==7].loc[df['year']==2013]
df_Coronation_July_2_1930 = df_Coronation_July_2_1930.set_index(df_Coronation_July_2_1930["tstamp"])
df_Coronation_July_2_1930.head(5)
```

Coronation Soap Opera Analysis for July 2nd at 19:00 GMT for 8.2M

```python
title = 'Coronation Soap Opera Analysis for July 2nd at 19:30 GMT for 8.2M'
```

```python
# Plot the day's energy consumption.

import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(12,4))
plt.plot(df_Coronation_July_2_1930["tstamp"],df_Coronation_July_2_1930["sigma"])
plt.xticks(rotation='vertical')
plt.title("Emmerdale Showing on July 6th at 19:00 GMT [5.03m]")
plt.xlabel("time stamp")
plt.ylabel("Energy Consumption (kW-h/hh")
plt.grid()
plt.plot(['19:00', '19:30'], [800, 800], color='r', linestyle='-', linewidth=2)
plt.show()
```

```python
# Create date series for each Acorn group.

# Acorn1 = A, B, C
# Acorn2 = D, E
# Acorn3 = F, G, H, I, J
# Acorn4 = K, L, M, N
# Acorn5 = O, P, Q

Acorn1  = df_Coronation_July_2_1930["A_sigma"] + df_Coronation_July_2_1930["B_sigma"] + df_Coronation_July_2_1930["C_sigma"]
Acorn2  = df_Coronation_July_2_1930["D_sigma"] + df_Coronation_July_2_1930["E_sigma"]
Acorn3  = df_Coronation_July_2_1930["F_sigma"] + df_Coronation_July_2_1930["G_sigma"] + df_Coronation_July_2_1930["H_sigma"] \
                                           + df_Coronation_July_2_1930["I_sigma"] + df_Coronation_July_2_1930["J_sigma"]
Acorn4  = df_Coronation_July_2_1930["K_sigma"] + df_Coronation_July_2_1930["L_sigma"] + df_Coronation_July_2_1930["M_sigma"] \
                                           + df_Coronation_July_2_1930["N_sigma"]
Acorn5  = df_Coronation_July_2_1930["O_sigma"] + df_Coronation_July_2_1930["P_sigma"] + df_Coronation_July_2_1930["Q_sigma"]
```

```python
# Plot the day's energy consumption for each Acorn group using a multiline plot.

plt.figure(figsize=(25,4))
hnd1, = plt.plot(df_Coronation_July_2_1930["tstamp"],Acorn1, label = 'Acorn 1')
hnd2, = plt.plot(df_Coronation_July_2_1930["tstamp"],Acorn2, label = 'Acorn 2')
hnd3, = plt.plot(df_Coronation_July_2_1930["tstamp"],Acorn3, label = 'Acorn 3')
hnd4, = plt.plot(df_Coronation_July_2_1930["tstamp"],Acorn4, label = 'Acorn 4')
hnd5, = plt.plot(df_Coronation_July_2_1930["tstamp"],Acorn5, label = 'Acorn 5')
plt.xticks(rotation='vertical')
plt.title("Power Usage per Acorn Group for showing of Coronation St on July 2nd at 19:30 GMT [5.03m]")
plt.xlabel("time stamp")
plt.ylabel("Energy Consumption (kW-h/hh")
plt.grid()
plt.plot(['19:30', '20:00'], [350, 350], color='r', linestyle='-', linewidth=2)
plt.legend(handles=[hnd1, hnd2, hnd3, hnd4, hnd5], loc="right")
plt.show()
```

```python
Acorn1 = Acorn1 / Acorn1['18:30']
Acorn2 = Acorn2 / Acorn2['18:30']
Acorn3 = Acorn3 / Acorn3['18:30']
Acorn4 = Acorn4 / Acorn4['18:30']
Acorn5 = Acorn5 / Acorn5['18:30']
```

```python
# Plot the day's normalized energy consumption for each Acorn group using a multiline plot.

plt.figure(figsize=(25,4))
hnd1, = plt.plot(df_Coronation_July_2_1930["tstamp"],Acorn1, label = 'Acorn 1')
hnd2, = plt.plot(df_Coronation_July_2_1930["tstamp"],Acorn2, label = 'Acorn 2')
hnd3, = plt.plot(df_Coronation_July_2_1930["tstamp"],Acorn3, label = 'Acorn 3')
hnd4, = plt.plot(df_Coronation_July_2_1930["tstamp"],Acorn4, label = 'Acorn 4')
hnd5, = plt.plot(df_Coronation_July_2_1930["tstamp"],Acorn5, label = 'Acorn 5')
plt.xticks(rotation='vertical')
plt.title("Normalized Power Usage per Acorn Group for showing of Coronation St on July 2nd at 19:30 GMT [5.03m]")
plt.xlabel("time stamp")
plt.ylabel("Normalized Energy Consumption (kW-h/hh")
plt.grid()
plt.plot(['19:30', '20:00'], [0.9, 0.9], color='r', linestyle='-', linewidth=2)
plt.legend(handles=[hnd1, hnd2, hnd3, hnd4, hnd5], loc="right")
plt.show()
```
